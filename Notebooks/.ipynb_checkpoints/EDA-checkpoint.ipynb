```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime
```

```python
cab=pd.read_csv('../data/Cab_Data.csv')
cab
```

```python
city=pd.read_csv('../data/City.csv')
city
```

```python
cust=pd.read_csv('../data/Customer_ID.csv')
cust
```

```python
trans=pd.read_csv('../data/Transaction_ID.csv')
trans
```

```python
df1=pd.merge(cab, city, on='City')
df1
```

```python
df2=pd.merge(df1, trans, on='Transaction ID')
df2
```

```python
df=pd.merge(df2, cust, on='Customer ID')
df
```

```python
df.isna().sum().sum()
```

```python
df['Users']=df['Users'].str.replace(',',"").astype(int)
df['Population']=df['Population'].str.replace(',',"").astype(int)
```

```python
df['Date of Travel']=pd.to_datetime(df['Date of Travel'])
```

```python
df['Day'] = df['Date of Travel'].dt.day
df['Month'] = df['Date of Travel'].dt.month
df['Year'] = df['Date of Travel'].dt.year
```

```python
df['Profit']=df['Price Charged']-df['Cost of Trip']
```

```python
df.dtypes
```

```python
df.columns
```

```python
reindex=['Transaction ID',
         'Customer ID',
         'Date of Travel',
         'Day',
         'Month',
         'Year',
         'Company',
         'City',
         'KM Travelled',
         'Price Charged',
         'Cost of Trip',
         'Profit',
         'Population',
         'Users',
         'Payment_Mode',
         'Gender',
         'Age',
         'Income (USD/Month)',
]
df=df.reindex(columns=reindex)
```

```python
df.describe()
```

```python
df.describe(exclude='number')
```

```python
df
```

```python
ydf=df[df['Company']=='Yellow Cab']
ydf
```

```python
pdf=df[df['Company']=='Pink Cab']
pdf
```

## EDA

#### Profit change per months over the years

```python
sns.relplot(x="Month", y="Profit", hue="Year",data=ydf, kind='line',ci=None)
sns.relplot(x="Month", y="Profit", hue="Year",data=pdf, kind='line',ci=None)
```

```python
sns.relplot(x="Month", y="Profit", hue="Year",data=pdf, kind='line',ci=None)
```

#### Customer Demographics based on City

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='City',data=ydf, hue='Gender')
plt.xticks(rotation=90)
plt.show()
plt.figure(figsize=(10, 6))
sns.countplot(x='City',data=pdf, hue='Gender')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(x='City',y='Age',data=ydf)
plt.xticks(rotation=90)
plt.show()
plt.figure(figsize=(10, 6))
sns.stripplot(x='City',y='Age',data=pdf)
plt.xticks(rotation=90)
plt.show()
```
