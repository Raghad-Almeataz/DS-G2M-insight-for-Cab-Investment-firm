```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime
```

```python
cab=pd.read_csv('../data/Cab_Data.csv')
cab
```

```python
city=pd.read_csv('../data/City.csv')
city
```

```python
cust=pd.read_csv('../data/Customer_ID.csv')
cust
```

```python
trans=pd.read_csv('../data/Transaction_ID.csv')
trans
```

```python
df1=pd.merge(cab, city, on='City')
df1
```

```python
df2=pd.merge(df1, trans, on='Transaction ID')
df2
```

```python
df=pd.merge(df2, cust, on='Customer ID')
df
```

## Cleaning


### Check NAs

```python
df.isna().sum().sum()
```

### Converting Users and Population columns into integers

```python
df['Users']=df['Users'].str.replace(',',"").astype(int)
df['Population']=df['Population'].str.replace(',',"").astype(int)
```

### Converting Date of Travel column into a date column

```python
df['Date of Travel']=pd.to_datetime(df['Date of Travel'])
```

### Extracting Days, Months, and Years out of Date of travel column to new columns

```python
df['Day'] = df['Date of Travel'].dt.day
df['Month'] = df['Date of Travel'].dt.month
df['Year'] = df['Date of Travel'].dt.year
```

### Calculating Profit by subtracting costs from revenue

```python
df['Profit']=df['Price Charged']-df['Cost of Trip']
```

### Convert Transaction ID and Customer ID columns into strings

```python
df['Transaction ID']=df['Transaction ID'].astype(str)
df['Customer ID']=df['Customer ID'].astype(str)
```

### Check Columns Datatypes

```python
df.dtypes
```

### Check column names and see how it is arranged

```python
df.columns
```

### Rearrange columns

```python
reindex=['Transaction ID',
         'Customer ID',
         'Date of Travel',
         'Day',
         'Month',
         'Year',
         'Company',
         'City',
         'KM Travelled',
         'Price Charged',
         'Cost of Trip',
         'Profit',
         'Population',
         'Users',
         'Payment_Mode',
         'Gender',
         'Age',
         'Income (USD/Month)',
]
df=df.reindex(columns=reindex)
```

## EDA

### Check dataset statistical information 

```python
# Continuous Data
df.describe()
```

```python
# Catigorical Data
df.describe(exclude='number')
```

## Insights

#### Profit change per months over the years

```python
a=sns.relplot(x="Month", y="Profit", hue="Year",data=df, kind='line',ci=None, col='Company')
a.fig.suptitle('Average profit change per months over the years', fontsize=12)
a.fig.subplots_adjust(top=0.8)
plt.show()
#plt.savefig('../Photos/ProfitChange.png')
```

#### Customer Retention

```python
fig, axes = plt.subplots(1,2, sharey=True, figsize=(14,5))
fig.suptitle('Customer Retention', fontsize=14)
df[df['Company']=='Yellow Cab']['Customer ID'].value_counts().plot(ax=axes[1])
axes[1].set_title('Yellow Cab')
axes[1].set_xlabel('Customer ID')
axes[0].set_ylabel('Number of transactions')
df[df['Company']=='Pink Cab']['Customer ID'].value_counts().plot(ax=axes[0])
axes[0].set_title('Pink Cab')
axes[0].set_xlabel('Customer ID')
plt.show()
#plt.savefig('../Photos/CustomerRetention.png')
```

#### Which cab is more used in each City

```python
b=sns.catplot(x='City',data=df, hue='Gender',kind="count",col='Company').set_xticklabels(rotation=90)
b.fig.suptitle('Which cab is more used in each City', fontsize=12)
b.fig.subplots_adjust(top=0.8)
plt.show()
#plt.savefig('../Photos/CabUsed.png')
```

#### Customer Demographics based on City

```python
c=sns.relplot(x='City',y='Age', hue='Gender',data=df,kind='line',ci=None, col='Company').set_xticklabels(rotation=90)
c.fig.suptitle('Average customer ages distributed by cities', fontsize=12)
c.fig.subplots_adjust(top=0.8)
plt.show()
#plt.savefig('../Photos/CustomerDemog1.png')
```

```python
d=sns.relplot(x='City',y='Income (USD/Month)', hue='Gender',data=df,kind='line',ci=None, col='Company').set_xticklabels(rotation=90)
d.fig.suptitle('Average customer income distributed by cities', fontsize=12)
d.fig.subplots_adjust(top=0.8)
plt.show()
#plt.savefig('../Photos/CustomerDemog2.png')
```

#### Which Cab is more expensive on customers?

```python
e=sns.relplot(x='KM Travelled',y='Price Charged',data=df,col='Company')
e.fig.suptitle('Which Cab is more expensive on customers?', fontsize=12)
e.fig.subplots_adjust(top=0.8)
plt.show()
#plt.savefig('../Photos/ExpCab.png')
```

#### Which Cab costs more?

```python
f=sns.relplot(y='Cost of Trip',x='Month',col='Company', data=df, kind='line', hue='Year', ci=None)
f.fig.suptitle('Which Cab costs more?', fontsize=12)
f.fig.subplots_adjust(top=0.8)
plt.show()
#plt.savefig('../Photos/CostCab.png')
```
