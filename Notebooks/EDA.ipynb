```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime
```

```python
cab=pd.read_csv('../data/Cab_Data.csv')
cab
```

```python
city=pd.read_csv('../data/City.csv')
city
```

```python
cust=pd.read_csv('../data/Customer_ID.csv')
cust
```

```python
trans=pd.read_csv('../data/Transaction_ID.csv')
trans
```

```python
df1=pd.merge(cab, city, on='City')
df1
```

```python
df2=pd.merge(df1, trans, on='Transaction ID')
df2
```

```python
df=pd.merge(df2, cust, on='Customer ID')
df
```

```python
df.isna().sum().sum()
```

```python
df['Users']=df['Users'].str.replace(',',"").astype(int)
df['Population']=df['Population'].str.replace(',',"").astype(int)
```

```python
df['Date of Travel']=pd.to_datetime(df['Date of Travel'])
```

```python
df['Day'] = df['Date of Travel'].dt.day
df['Month'] = df['Date of Travel'].dt.month
df['Year'] = df['Date of Travel'].dt.year
```

```python
df['Profit']=df['Price Charged']-df['Cost of Trip']
```

```python
df['Transaction ID']=df['Transaction ID'].astype(str)
df['Customer ID']=df['Customer ID'].astype(str)
```

```python
df.dtypes
```

```python
df.columns
```

```python
reindex=['Transaction ID',
         'Customer ID',
         'Date of Travel',
         'Day',
         'Month',
         'Year',
         'Company',
         'City',
         'KM Travelled',
         'Price Charged',
         'Cost of Trip',
         'Profit',
         'Population',
         'Users',
         'Payment_Mode',
         'Gender',
         'Age',
         'Income (USD/Month)',
]
df=df.reindex(columns=reindex)
```

## EDA

#### Profit change per months over the years

```python
sns.relplot(x="Month", y="Profit", hue="Year",data=df, kind='line',ci=None, col='Company')
plt.show()
```

#### Customer Demographics based on City

```python
sns.catplot(x='City',data=df, hue='Gender',kind="count",col='Company').set_xticklabels(rotation=90)
plt.show()
```

```python
sns.relplot(x='City',y='Age', hue='Gender',data=df,kind='line',ci=None, col='Company').set_xticklabels(rotation=90)
plt.show()
```

```python
sns.relplot(x='City',y='Income (USD/Month)', hue='Gender',data=df,kind='line',ci=None, col='Company').set_xticklabels(rotation=90)
plt.show()
```

#### Which Cab is more expensive on customers?

```python
sns.relplot(x='Price Charged',y='KM Travelled',data=df,col='Company')
plt.show()
```

#### Which Cab costs more?

```python
sns.relplot(x='Cost of Trip',y='Profit',col='Company',data=df)
plt.show()
```

#### Which cab is more used in each City

```python
ydf.groupby('City')['Transaction ID'].count().plot(kind='bar')
plt.show()
pdf.groupby('City')['Transaction ID'].count().plot(kind='bar')
plt.show()
```
