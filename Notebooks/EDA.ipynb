```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime
```

```python
cab=pd.read_csv('../data/Cab_Data.csv')
cab
```

```python
city=pd.read_csv('../data/City.csv')
city
```

```python
cust=pd.read_csv('../data/Customer_ID.csv')
cust
```

```python
trans=pd.read_csv('../data/Transaction_ID.csv')
trans
```

```python
df1=pd.merge(cab, city, on='City')
df1
```

```python
df2=pd.merge(df1, trans, on='Transaction ID')
df2
```

```python
df=pd.merge(df2, cust, on='Customer ID')
df
```

```python
df.isna().sum().sum()
```

```python
df['Users']=df['Users'].str.replace(',',"").astype(int)
df['Population']=df['Population'].str.replace(',',"").astype(int)
```

```python
df['Date of Travel']=pd.to_datetime(df['Date of Travel'])
```

```python
df['Day'] = df['Date of Travel'].dt.day
df['Month'] = df['Date of Travel'].dt.month
df['Year'] = df['Date of Travel'].dt.year
```

```python
df['Profit']=df['Price Charged']-df['Cost of Trip']
```

```python
df['Transaction ID']=df['Transaction ID'].astype(str)
df['Customer ID']=df['Customer ID'].astype(str)
```

```python
df.dtypes
```

```python
df.columns
```

```python
reindex=['Transaction ID',
         'Customer ID',
         'Date of Travel',
         'Day',
         'Month',
         'Year',
         'Company',
         'City',
         'KM Travelled',
         'Price Charged',
         'Cost of Trip',
         'Profit',
         'Population',
         'Users',
         'Payment_Mode',
         'Gender',
         'Age',
         'Income (USD/Month)',
]
df=df.reindex(columns=reindex)
```

## EDA

#### Profit change per months over the years

```python
a=sns.relplot(x="Month", y="Profit", hue="Year",data=df, kind='line',ci=None, col='Company')
a.fig.suptitle('Average profit change per months over the years', fontsize=12)
a.fig.subplots_adjust(top=0.8)
plt.show()
```

#### Which cab is more used in each City

```python
b=sns.catplot(x='City',data=df, hue='Gender',kind="count",col='Company').set_xticklabels(rotation=90)
b.fig.suptitle('Which cab is more used in each City', fontsize=12)
b.fig.subplots_adjust(top=0.8)
plt.show()
```

### Customer Demographics based on City

```python
c=sns.relplot(x='City',y='Age', hue='Gender',data=df,kind='line',ci=None, col='Company').set_xticklabels(rotation=90)
c.fig.suptitle('Average customer ages distributed by cities', fontsize=12)
c.fig.subplots_adjust(top=0.8)
plt.show()
```

```python
d=sns.relplot(x='City',y='Income (USD/Month)', hue='Gender',data=df,kind='line',ci=None, col='Company').set_xticklabels(rotation=90)
d.fig.suptitle('Average customer income distributed by cities', fontsize=12)
d.fig.subplots_adjust(top=0.8)
plt.show()
```

#### Which Cab is more expensive on customers?

```python
e=sns.relplot(x='Price Charged',y='KM Travelled',data=df,col='Company')
e.fig.suptitle('Which Cab is more expensive on customers?', fontsize=12)
e.fig.subplots_adjust(top=0.8)
plt.show()
```

#### Which Cab costs more?

```python
f=sns.relplot(x='Cost of Trip',y='Profit',col='Company',data=df)
f.fig.suptitle('Which Cab costs more?', fontsize=12)
f.fig.subplots_adjust(top=0.8)
plt.show()
```

#### Which cab is more used in each City

```python
figure, axes = plt.subplots(1, 2, sharey=True,figsize=(12,6),sharex=True)
figure.suptitle('Which cab is more used in each city?')
axes[0].set_title('Yellow Cab')
df[df['Company']=='Yellow Cab'].groupby('City')['Transaction ID'].count().plot.bar(ax=axes[0])
axes[1].set_title('Pink Cab')
df[df['Company']=='Pink Cab'].groupby('City')['Transaction ID'].count().plot.bar(ax=axes[1])
axes[0].set_ylabel('Number of transactions per city')
plt.show()
```
